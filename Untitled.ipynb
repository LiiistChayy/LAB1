<h1 style="color: #2E86C1;">Постановка задачи, условие:</h1>

На складах $A_1$ и $A_2$ содержится сталь двух марок: «а» и «б».  
Сталь необходимо доставить в пункты потребления $B_1$ и $B_2$.

Известно, что сталь разных марок может заменяться:  
$1$ т марки «а» заменяет $0.7$ т марки «б». Обратная замена невозможна.

За каждую перевозку установлена стоимость в денежных единицах, зависящая от маршрута.

**Цель:** найти план перевозок, минимизирующий общие затраты.

# <h1 style="color: #2E86C1;">Формальная постановка:</h1>
## Переменные

- $x_{ij}^a$ — количество стали марки «а», перевезённое со склада $i$ в пункт $j$.
- $x_{ij}^b$ — количество стали марки «б», перевезённое со склада $i$ в пункт $j$.
- $s_{ij}$ — часть $x_{ij}^a$, направленная на замену стали «б» в пункте $j$.

где $i = 1,2$ (склады $A_1, A_2$), $j = 1,2$ (пункты $B_1, B_2$).

## Целевая функция

Минимизация затрат:

$$
Z = \sum_{i=1}^{2}\sum_{j=1}^{2} c_{ij}\bigl(x_{ij}^a + x_{ij}^b\bigr) + \Delta \sum_{i=1}^{2}\sum_{j=1}^{2} x_{ij}^b \to \min
$$

с тарифами:  
$c_{11}=1.5,\; c_{12}=1.7,\; c_{21}=1.4,\; c_{22}=1.7$ (ДЕ/т).  
Параметр $\Delta$ используется в параметрическом анализе.

## Ограничения

### Запасы на складах
- $A_1$:  
  $x_{11}^a + x_{12}^a \le 3600$  
  $x_{11}^b + x_{12}^b \le 5800$
- $A_2$:  
  $x_{21}^a + x_{22}^a \le 5000$  
  $x_{21}^b + x_{22}^b \le 2400$

### Потребности в пункте $B_1$
- $ (x_{11}^a - s_{11}) + (x_{21}^a - s_{21}) \ge 1100$
- $ (x_{11}^b + x_{21}^b) + 0.7(s_{11}+s_{21}) \ge 4800$
- $ (x_{11}^a + x_{11}^b) + (x_{21}^a + x_{21}^b) = 6600$

### Потребности в пункте $B_2$
- $ (x_{12}^a - s_{12}) + (x_{22}^a - s_{22}) \ge 4500$
- $ (x_{12}^b + x_{22}^b) + 0.7(s_{12}+s_{22}) \ge 2200$
- $ (x_{12}^a + x_{12}^b) + (x_{22}^a + x_{22}^b) = 7200$

### Связь переменных
- $s_{ij} \le x_{ij}^a \quad \forall i,j$
- $x_{ij}^a, x_{ij}^b, s_{ij} \ge 0$

### Инструменты: Python + Pulp + Jupyter

````python
import pulp
````

````python
def solve_steel_problem(a1_a, a1_b, a2_a, a2_b, cost_b_mod=0):
    prob = pulp.LpProblem("Steel_Optimization", pulp.LpMinimize)
    warehouses = [1, 2]; points = [1, 2]
    x = pulp.LpVariable.dicts("x", (warehouses, points, ['a', 'b']), lowBound=0)
    s = pulp.LpVariable.dicts("subst", (warehouses, points), lowBound=0)
    costs = {1: {1: 1.5, 2: 1.7}, 2: {1: 1.4, 2: 1.7}}

    prob += pulp.lpSum([
        x[i][j]['a'] * costs[i][j] + x[i][j]['b'] * (costs[i][j] + cost_b_mod)
        for i in warehouses for j in points
    ])


    prob += pulp.lpSum([x[1][j]['a'] for j in points]) <= a1_a
    prob += pulp.lpSum([x[1][j]['b'] for j in points]) <= a1_b
    prob += pulp.lpSum([x[2][j]['a'] for j in points]) <= a2_a
    prob += pulp.lpSum([x[2][j]['b'] for j in points]) <= a2_b


    prob += pulp.lpSum([x[i][1]['a'] - s[i][1] for i in warehouses]) >= 1100
    prob += pulp.lpSum([x[i][1]['b'] + 0.7 * s[i][1] for i in warehouses]) >= 4800
    prob += pulp.lpSum([x[i][1]['a'] + x[i][1]['b'] for i in warehouses]) == 6600

    prob += pulp.lpSum([x[i][2]['a'] - s[i][2] for i in warehouses]) >= 4500
    prob += pulp.lpSum([x[i][2]['b'] + 0.7 * s[i][2] for i in warehouses]) >= 2200
    prob += pulp.lpSum([x[i][2]['a'] + x[i][2]['b'] for i in warehouses]) == 7200

    for i in warehouses:
        for j in points:
            prob += s[i][j] <= x[i][j]['a']

    prob.solve(pulp.PULP_CBC_CMD(msg=0))
    ts = sum(pulp.value(s[i][j]) for i in warehouses for j in points)
    return prob, ts
````

````python
a1a, a1b, a2a, a2b = 3600, 5800, 5000, 2400
res, ts = solve_steel_problem(a1a, a1b, a2a, a2b)

print("--- Базовое решение ---")
print(f"Общая стоимость: {pulp.value(res.objective)} ДЕ")
print(f"Объем замены марки 'а' на 'б': {ts} тонн")
````

````python
print("--- Исследование 1.1 ---")
while ts < 100:
    a1b -= 50; a2b -= 50  
    a1a += 50; a2a += 50  
    res, ts = solve_steel_problem(a1a, a1b, a2a, a2b)
    if a1b < 0 or a2b < 0: break

print(f"Результат: замена в {round(ts, 2)} т появилась при запасах 'б': А1={a1b}, А2={a2b}")
````

````python
print("--- Исследование 2.1 ---")
a1a, a1b, a2a, a2b = 10000, 10000, 10000, 10000 
mod = 0
res, ts = solve_steel_problem(a1a, a1b, a2a, a2b, cost_b_mod=mod)

while ts < 100:
    mod += 0.1 
    res, ts = solve_steel_problem(a1a, a1b, a2a, a2b, cost_b_mod=mod)

print(f"Результат: замена в {round(ts, 2)} т появилась при наценке на марку 'б' = {round(mod, 2)} ДЕ")
````
